```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Hyperparameters
BATCH_SIZE = 32
NUM_EPOCHS = 20
LEARNING_RATE = 0.001
NUM_CLASSES = 2
IMAGE_SIZE = 224
```

```python
# Data transforms
transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
])

# Updated dataset paths
train_dataset = datasets.ImageFolder(root='/content/drive/MyDrive/drone_detection/data/train', transform=transform)
val_dataset = datasets.ImageFolder(root='/content/drive/MyDrive/drone_detection/data/val', transform=transform)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)
```

```python
# CNN model
class DroneCNN(nn.Module):
    def __init__(self, num_classes=NUM_CLASSES):
        super(DroneCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, stride=1, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.pool = nn.MaxPool2d(2, 2)

        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.bn2 = nn.BatchNorm2d(64)

        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1)
        self.bn3 = nn.BatchNorm2d(128)

        self.conv4 = nn.Conv2d(128, 256, kernel_size=3, stride=1, padding=1)
        self.bn4 = nn.BatchNorm2d(256)

        self.conv5 = nn.Conv2d(256, 512, kernel_size=3, stride=1, padding=1)
        self.bn5 = nn.BatchNorm2d(512)

        self.fc1 = nn.Linear(512 * 7 * 7, 128)
        self.dropout = nn.Dropout(0.5)
        self.fc2 = nn.Linear(128, num_classes)

        self.leaky_relu = nn.LeakyReLU(0.1)

    def forward(self, x):
        x = self.pool(self.leaky_relu(self.bn1(self.conv1(x))))
        x = self.pool(self.leaky_relu(self.bn2(self.conv2(x))))
        x = self.pool(self.leaky_relu(self.bn3(self.conv3(x))))
        x = self.pool(self.leaky_relu(self.bn4(self.conv4(x))))
        x = self.pool(self.leaky_relu(self.bn5(self.conv5(x))))

        x = x.view(-1, 512 * 7 * 7)
        x = self.leaky_relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = DroneCNN().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
```

```python
# Training loop
for epoch in range(NUM_EPOCHS):
    model.train()
    running_loss = 0
    correct = 0
    total = 0

    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    train_loss = running_loss / total
    train_acc = correct / total

    model.eval()
    val_loss = 0
    val_correct = 0
    val_total = 0

    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            val_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            val_correct += (preds == labels).sum().item()
            val_total += labels.size(0)

    val_loss /= val_total
    val_acc = val_correct / val_total

    print(f'Epoch {epoch+1}/{NUM_EPOCHS} - '
          f'Train Loss: {train_loss:.4f} - Train Acc: {train_acc:.4f} - '
          f'Val Loss: {val_loss:.4f} - Val Acc: {val_acc:.4f}')
```

```python
from tqdm import tqdm

for epoch in range(NUM_EPOCHS):
    model.train()
    running_loss = 0
    correct = 0
    total = 0

    # Progress bar for training batches
    train_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{NUM_EPOCHS} Training", leave=False)
    for inputs, labels in train_bar:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

        train_loss = running_loss / total
        train_acc = correct / total
        train_bar.set_postfix(loss=train_loss, accuracy=train_acc)

    model.eval()
    val_loss = 0
    val_correct = 0
    val_total = 0

    # Progress bar for validation batches
    val_bar = tqdm(val_loader, desc=f"Epoch {epoch+1}/{NUM_EPOCHS} Validation", leave=False)
    with torch.no_grad():
        for inputs, labels in val_bar:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            val_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            val_correct += (preds == labels).sum().item()
            val_total += labels.size(0)

            val_loss_epoch = val_loss / val_total
            val_acc = val_correct / val_total
            val_bar.set_postfix(loss=val_loss_epoch, accuracy=val_acc)

    print(f'Epoch {epoch+1}/{NUM_EPOCHS} - '
          f'Train Loss: {train_loss:.4f} - Train Acc: {train_acc:.4f} - '
          f'Val Loss: {val_loss_epoch:.4f} - Val Acc: {val_acc:.4f}',
          flush=True)
```

```python
import torch
import torch.nn as nn

# Load your model and weights
model = DroneCNN()
model.load_state_dict(torch.load('/content/drive/MyDrive/drone_detection/drone_cnn_fpga.pth'))
model.eval()

def count_params(layer):
    return sum(p.numel() for p in layer.parameters())

def size_in_bytes(num_params, bits=32):
    return num_params * bits // 8

def macs_conv2d(layer, input_size):
    # MACs = output H * output W * kernel H * kernel W * in_channels * out_channels
    kernel_h, kernel_w = layer.kernel_size
    in_channels = layer.in_channels
    out_channels = layer.out_channels
    stride = layer.stride[0]
    padding = layer.padding[0]

    out_h = (input_size[0] - kernel_h + 2 * padding) // stride + 1
    out_w = (input_size[1] - kernel_w + 2 * padding) // stride + 1
    return out_h * out_w * kernel_h * kernel_w * in_channels * out_channels

def macs_linear(layer):
    # MACs = input features * output features
    return layer.in_features * layer.out_features

input_size = (224, 224)  # Initial image size (HxW)

print("Layer-wise parameter count, memory and estimated MACs\n")
for name, layer in model.named_modules():
    if isinstance(layer, nn.Conv2d):
        params = count_params(layer)
        mem = size_in_bytes(params)
        macs = macs_conv2d(layer, input_size)
        print(f"{name} - Conv2d: Params={params}, Memory={mem} bytes, MACs={macs}")
        # Update input size after conv + pooling (assuming 2x2 max pool)
        input_size = (input_size[0] // 2, input_size[1] // 2)
    elif isinstance(layer, nn.Linear):
        params = count_params(layer)
        mem = size_in_bytes(params)
        macs = macs_linear(layer)
        print(f"{name} - Linear: Params={params}, Memory={mem} bytes, MACs={macs}")

# Total model parameters and memory
total_params = count_params(model)
total_mem = size_in_bytes(total_params)

print(f"\nTotal parameters: {total_params}")
print(f"Approximate model memory size: {total_mem} bytes (32-bit float)")
```

```python
import torch
import os
from torchvision import datasets, transforms
from tqdm import tqdm

# Path to quantized model's state_dict
quantized_weights_path = '/content/drive/MyDrive/drone_detection/drone_cnn_fpga_quantized.pth'

file_size_mb = os.path.getsize(quantized_weights_path) / (1024*1024)
print(f"Quantized weights file size: {file_size_mb:.2f} MB")

# Create model instance and quantize it (to match saved state dict)
model = DroneCNN()
model.eval()

quantized_model = torch.quantization.quantize_dynamic(
    model,
    {torch.nn.Linear, torch.nn.Conv2d},
    dtype=torch.qint8)

# Load quantized weights state dict
quantized_model.load_state_dict(torch.load(quantized_weights_path))

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
quantized_model.to(device)
quantized_model.eval()

# Setup validation data loader
transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
])
val_dataset = datasets.ImageFolder('/content/drive/MyDrive/drone_detection/data/val', transform=transform)
val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=32, shuffle=False)

criterion = torch.nn.CrossEntropyLoss()
running_loss = 0.0
correct = 0
total = 0

with torch.no_grad():
    for inputs, labels in tqdm(val_loader, desc="Evaluating Quantized Model"):
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = quantized_model(inputs)
        loss = criterion(outputs, labels)
        running_loss += loss.item() * inputs.size(0)
        _, predicted = torch.max(outputs, 1)
        correct += (predicted == labels).sum().item()
        total += labels.size(0)

val_loss = running_loss / total
val_accuracy = correct / total
val_error = 1 - val_accuracy

print(f"\nValidation Loss: {val_loss:.4f}")
print(f"Validation Accuracy: {val_accuracy:.4f}")
print(f"Validation Error: {val_error:.4f}")
```
